```python
# more nos of plots in a graph

x = np.linspace(10,20,50)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.figure(figsize=((10,5)))
plt.plot(x,y,c='r',marker='o',markersize=5,linestyle="-.",label="Sin") 
plt.scatter(x,z,c='g',marker='+',s=50,label="Cos")      # size - s 
plt.xlabel("linear values",fontsize=12)
plt.ylabel("Sin & Cos values",fontsize=12)
plt.title("Scatter Graph plot",fontsize=20)
plt.legend()                                    ### plt.legend is required for label to be written
plt.show()
```

```python
plt.figure(figsize=((10,5)))
plt.plot(x,y,c='r',marker='o',markersize=5,linestyle="-.",label="Sin") 
plt.scatter(x,z,c='g',marker='+',s=50,label="Cos")     
plt.xlabel("linear values",fontsize=12)
plt.ylabel("Sin & Cos values",fontsize=12)
plt.title("Scatter Graph plot",fontsize=20)
plt.xticks([10,15,20])      # only these values will be visible
plt.yticks([-1,-0.5,0,0.5,1])
plt.legend()
plt.show()
```

# Bargraph 

```python
usn = [f"4VP20AI0{v}" for v in range(10,15)]
```

```python
names = []
for i in range(10,15):
    s = ''
    for j in range(10):
        s = s + chr(97+np.random.randint(26))
    names.append(s)
names
```

```python
marks = [np.random.randint(60,100) for i in range(10,15)]
```

```python
for i in zip(usn,names,marks):
    print(i)
```

```python
plt.bar(usn,marks,color='g',width=1)           # here width = 1
plt.title("Marks barchart")
plt.xlabel("USN")
plt.ylabel("Marks")
plt.yticks([50,60,70,80,90,100])
plt.show()
```

```python
plt.bar(usn,marks,color='g',width=0.6)        # here width = 0.6
plt.title("Marks barchart")
plt.xlabel("USN")
plt.ylabel("Marks")
plt.yticks([50,60,70,80,90,100])
plt.show()
```

```python
# different colors to each 
l = plt.bar(usn,marks,color='g',width=0.6)

l[-1].set_color("r")
l[-2].set_color("b")
l[0].set_color("y")

plt.title("Marks barchart")
plt.xlabel("USN")
plt.ylabel("Marks")
plt.yticks([50,60,70,80,90,100])
plt.show()
```

```python
# instead of bar we can use barh for horizontal

# different ->  barh
l = plt.barh(usn,marks,color='g')

l[-1].set_color("r")
l[-2].set_color("b")
l[0].set_color("y")

plt.title("Marks barchart")
plt.xlabel("USN")
plt.ylabel("Marks")
plt.show()
```
