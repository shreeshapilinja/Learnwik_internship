Gradient Descent Algorithm <br>
- optimization algorithm
i.e first take a random line <br>
then take gradient descent for that <br>
then we get the optimal line (y = mx + x) by changing m , c

- while this 3 rd step we calculate the error and if error we update the weight <br>
w = w - n <sub>^</sub>w  <br>
<sub>^</sub>w = del(loss or error ) / w <br>

running this some times i.e eg 100 epochs
and having some learning rate n

- learning rate is nothing but step size 
- if larger learning rate then we will not get the optimal weight will be either increasing or decreasing 
- if smaller learning rate it gives optimal but takes more time 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([1.5,3.2,7,7.5,9,11.9,15,15.8])
```

```python
plt.plot(x,y,marker="o")
plt.show()
```

```python
# 1) take a random line i.e m and c value 

# w1 = m  , w0 = c

w1,w0 = np.random.rand(2)
```

```python
# a random line we get 

w1,w0
```

```python
# now using weights adjusting get a line by removing the error 
# w1*xi+w0   i.e y = mx+c 

# 2) generate a random line 
y_hat = []
for xi in x:
    y_hat.append(w1*xi+w0)
y_hat
```

```python
plt.plot(x,y,marker="o",label="actual line")
plt.plot(x,y_hat,c='r',marker="+",label="random line")
plt.legend()
plt.show()
```

```python
y
```

```python
y_hat = np.array(y_hat)
y_hat
```

```python
squareerror = sum((y-y_hat)**2)
squareerror
```

```python
# learning rate - lr  min 

lr = 0.001
epochs = 1000
error = []
```

```python
# gradient descent 

for ep in range(epochs):
    ep_error , dw1,dw0 = 0,0,0
    
    # loss in current epoch
    for i in range(len(x)):
        y_pred = w1*x[i]+w0
        ep_error += (y[i]-y_pred)**2
    error.append(ep_error)
    
    # calculation of partial derivative (dw1,dw0)
    for i in range(len(x)):
        dw1 += -2*x[i]*(y[i]-(w1*x[i]+w0))  # 2x(y-y_hat)
        dw0 += -2*(y[i]-(w1*x[i]+w0))  # 2(y-y_hat)
    
    # updating the weights 
    w1 = w1 - lr*dw1
    w0 = w0 - lr*dw0
    print(f"w1={w1} , w0={w0}")
```

```python
# now using weights adjusted to get a line by removing the error 
y_pred = []
for xi in x:
    y_pred.append(w1*xi+w0)
y_pred
```

```python
plt.plot(x,y,marker="o",label="actual line")
plt.plot(x,y_pred,c='r',marker="+",label="predicted line")
plt.legend()
plt.show()
```

```python
# we can observe the random line moved near to actual line 

plt.plot(x,y,marker="o",label="actual line")
plt.plot(x,y_hat,c='g',marker="*",label="random line")
plt.plot(x,y_pred,c='r',marker="+",label="predicted line")
plt.legend()
plt.show()
```
